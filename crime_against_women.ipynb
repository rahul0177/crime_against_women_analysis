```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('crimes_against_women_2001-2014.csv')
dataframe=df.head()
dataframe
```

FROM above analysis it is cleared that there is no null value present in dataset. so from here it is cleared that dataset is ready to analysis.

```python
district_count=df['STATE/UT'].value_counts()
district_count=district_count[:35]
district_count
```

This show counts of district in each state in descending order where more crime against women is taken

```python
district_count.plot.pie(radius=3,autopct='%.1f%%',shadow=True)
```

From the Pie chart it is cleared that no.of crime against women is maximum in himanchal pradesh and uttarpradesh, and no. of crime against women is minimum in D&N Haveli ,Lakshadweep , A&N Islands ,Chandigarh,A & N Islands 

```python
STATE_IN_INDIA=df['STATE/UT'].unique()
STATE_IN_INDIA=STATE_IN_INDIA[:35]
STATE_IN_INDIA
```

```python
types_of_crime=list(df.columns.values)
types_of_crime=types_of_crime[4:]
types_of_crime
```

```python
crime_against_women=df[types_of_crime].sum()
crime_against_women
```

```python
crime_against_women.plot.barh()
```

FROM ABOVE BAR GRAPH IT IS CLEARED THAT CRIME LIKE 'Cruelty by Husband or his Relatives' is high and there is no case of crime of kind 'IMPORTION OF GIRLS'. 

STATE WISE CRIME ANALYSIS:-

```python
for state in STATE_IN_INDIA:
    fig=plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
    plt.title(state)
    plt.xlabel('crimes')
    plt.ylabel('No of Cases')
    stt=df[(df['STATE/UT']==state)]
    crm=stt.iloc[:,4:].sum()
    crm.plot.barh()
        
```
